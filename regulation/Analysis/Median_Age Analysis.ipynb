```python
# import dependencies
from path import Path
import pandas as pd
import numpy as np
import pandas as pd
from pandas import Series

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# Import dataset & store into dataframe
merged_covid_df = pd.read_csv("../Resources/merged_covid_FINAL.csv")
merged_covid_df.head()
```

```python
# Import dataset & store into dataframe
owid_df = pd.read_csv("../Resources/raw/owid-covid-data(Aug31,2020).csv")
owid_df.head()
```

```python
updated_owid_df=owid_df
updated_owid_df.head()
```

```python
covid_owid_df=updated_owid_df.groupby("continent").mean()["extreme_poverty"]
covid_owid_df.head()
```

```python
# Create the plot.
covid_owid_df.plot( marker="*", color="green", linewidth=2, label='Extreme Poverty')
# Create labels for the x and y axes.
plt.xlabel("Continent")
plt.ylabel("Extreme Poverty Rating")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Poverty Level Per Continent")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.figure(figsize=(80, 100))
```

```python
#Dataframe for Canada only

merged_covid_canada_df = pd.DataFrame(merged_covid_df.loc[merged_covid_df["location"]=="Canada"].reset_index())
merged_covid_canada_df = merged_covid_canada_df.drop(["index"],axis=1)
merged_covid_canada_df.head()
```

## AFRICA

```python
#  Dataframe filtered for Continent
merged_covid_AF_df = pd.DataFrame(merged_covid_df.loc[merged_covid_df["continent"]=="Africa"].reset_index())
merged_covid_AF_df = merged_covid_AF_df.drop(["index"],axis=1)
merged_covid_AF_df
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(10,10))
ax.barh(merged_covid_AF_df["location"],merged_covid_AF_df["median_age"])

ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(5))
plt.show()
```

```python
# Calculate the average test scores.
covid_Africa_df = merged_covid_AF_df.groupby(["location"]).mean()["median_age"]
covid_Africa_df
```

```python
len(covid_Africa_df)
```

```python
# Median age range for different locations
bins = [0,20,30,40,50]
median_age_Africa=covid_Africa_df.groupby(pd.cut(covid_Africa_df,bins)).count()
median_age_Africa
```

```python
# Create the plot.
median_age_Africa.plot( marker="*", color="green", linewidth=2, label='Median Age')
# Create labels for the x and y axes.
plt.xlabel("Median Age")
plt.ylabel("Count")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Covid 19 Impacted Median Age Group in Sub Saharan Africa")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.figure(figsize=(80, 100))
```

```python
# Establish the  bins and group names.
bins = [0, 20, 30,40,50]
group_names = ["<19","20-29","30-39","40-50"]
group_names
```

```python
median_age_Africa=merged_covid_AF_df.groupby("location").mean()
median_age_Africa.head()
```

```python
# Categorize Age based on the bins.
median_age_Africa[("Age_Range")] = pd.cut(covid_Africa_df,bins, labels=group_names)

median_age_Africa.head()
```

```python
fig, ax = plt.subplots()
#ax2 = ax.twinx()
ax.bar(median_age_Africa.Age_Range, median_age_Africa["total_cases_updated"], color="red", label="total_deaths_updated")
#ax2.plot(median_age_Asia_df.Age_Range, median_age_Asia_df["total_cases_updated"], color='blue', label='total_cases_updated')
#ax.set_xticklabels(median_age_Asia_df.Age_Range)
#ax.legend(loc='best')
ax.set_ylabel('Total Cases Per Age Group in Asia', color="red")
#ax2.set_ylabel('total deaths', color="blue")
```

```python
# Create the plot.
plt.bar(median_age_Africa.Age_Range, median_age_Africa["total_cases_updated"], color="blue", label='Total Cases Updated')
# Create labels for the x and y axes.
plt.xlabel("Age Range")
plt.ylabel("Total Cases")
# Create a title.
plt.title("Total Cases Per Age Group in Africa")
# Add the legend.
plt.legend()
```
